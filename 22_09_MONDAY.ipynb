```python
#spirit servings

q1 = np.quantile(beer['spirit_servings'],0.25)
q3 = np.quantile(beer['spirit_servings'],0.75)
iqr = q3 - q1
upper_bound = q3 + (1.5 * iqr)
lower_bound = q1 - (1.5 * iqr)

print(upper_bound)
print(lower_bound)
print(iqr)
```

```python
spirit_outliers = []
for i in beer['spirit_servings']:
    if i > upper_bound or i < lower_bound:
        spirit_outliers.append(i)

spirit_outliers
```

```python
beer['spirit_servings'] = beer['spirit_servings'].clip(lower=lower_bound,upper=upper_bound)
```

```python
plt.boxplot(beer['spirit_servings'])
```

```python
hr1['spirit_servings'] = hr1['spirit_servings'].clip(lower=lower_bound,upper=upper_bound)
plt.boxplot(hr1['spirit_servings'].dropna())
plt.title('spirit_servings after clipping')
plt.show()
```
