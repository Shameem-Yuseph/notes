```python
import requests
import pandas as pd

# Fetch data from API
url = "https://disease.sh/v3/covid-19/countries"
response = requests.get(url)
data = response.json()

# Load into DataFrame
covid_df = pd.DataFrame(data)

# Select only necessary columns
selected_columns = [
    "country", "cases", "todayCases",
    "deaths", "todayDeaths", "recovered",
    "active", "critical"
]
covid_selected = covid_df[selected_columns]

# Save to Excel
covid_selected.to_excel("covid_data.xlsx", index=False)

print("COVID-19 data has been saved to covid_data.xlsx")
```
